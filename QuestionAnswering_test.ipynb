```python
from collections import Counter
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
from transformers import T5Tokenizer, TFT5ForConditionalGeneration
```

```python
t5_model = TFT5ForConditionalGeneration.from_pretrained('t5-base')
t5_tokenizer = T5Tokenizer.from_pretrained('t5-base')

t5_model.summary()
```

"Use \<extra_id_0\>" special token as masked word modeling.  To construct sentences in fill in the blank format. 

```python
PROMPT_SENTENCE = ( "A poodle is a type of <extra_id_0> .")
t5_input_text = PROMPT_SENTENCE
t5_inputs = t5_tokenizer([t5_input_text], return_tensors='tf')
t5_summary_ids = t5_model.generate(t5_inputs['input_ids'],
                                   num_beams=9,
                                   no_repeat_ngram_size=1,
                                   num_return_sequences=3,
                                   min_length=1,
                                   max_length=5)

print([t5_tokenizer.decode(g, skip_special_tokens=True,
                           clean_up_tokenization_spaces=False) for g in t5_summary_ids])
```

```python

#Crafting a sentence
countries = ['England', 'France', 'Germany', 'Russia', 'Egypt', 'Thailand', 'Japan', 'Canada', 'India', 'China']
PROMPT_SENTENCE = ( f"The capital city of COUNTRY is <extra_id_0>.")

for country in countries:

  #t5_input_text = PROMPT_SENTENCE
  t5_input_text = PROMPT_SENTENCE.replace("COUNTRY", country)
  t5_inputs = t5_tokenizer([t5_input_text], return_tensors='tf')
  t5_summary_ids = t5_model.generate(t5_inputs['input_ids'],
                                   num_beams=9,
                                   no_repeat_ngram_size=2,
                                   num_return_sequences=3,
                                   min_length=1,
                                   max_length=3)

  print([t5_tokenizer.decode(g, skip_special_tokens=True,
                           clean_up_tokenization_spaces=False) for g in t5_summary_ids])
```
